```python
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl
```

## Membership Functions

```python
t_amb = ctrl.Antecedent(np.arange(0, 51, 1), 'Ambient temperature C')
t_amb['very_cold'] = fuzz.trimf(t_amb.universe, [0,0,16])
t_amb['cold'] = fuzz.trimf(t_amb.universe, [12,15,18])
t_amb['cool'] = fuzz.trimf(t_amb.universe, [16,19,22])
t_amb['neutral'] = fuzz.trimf(t_amb.universe, [20,23,26])
t_amb['warm'] = fuzz.trimf(t_amb.universe, [23,27,30])
t_amb['hot'] = fuzz.trimf(t_amb.universe, [27,30,33])
t_amb['very_hot'] = fuzz.trimf(t_amb.universe, [30,50,50])
t_amb.view()
```

```python
t_skin = ctrl.Antecedent(np.arange(0, 51, 1), 'Skin temperature C')
t_skin['cold_nociceptor'] = fuzz.trapmf(t_skin.universe, [0,0,10,14])
t_skin['vasocontstriction'] = fuzz.trapmf(t_skin.universe, [10,14,30,33])
t_skin['homeosthasis'] = fuzz.trapmf(t_skin.universe, [31,33,36,37])
t_skin['vasodilation'] = fuzz.trapmf(t_skin.universe, [34,38,40,45])
t_skin['heat_nociceptor'] = fuzz.trapmf(t_skin.universe, [39,44,50,50])
t_skin.view()
```

```python
force = ctrl.Antecedent(np.arange(0, 101, 1), 'Interfacing force [n]')
force['low'] = fuzz.trapmf(force.universe, [0,0,5,15])
force['moderate'] = fuzz.trapmf(force.universe, [5,15,30,40])
force['high'] = fuzz.trapmf(force.universe, [30,40,100,100])
force.view()
```

```python
comfort = ctrl.Consequent(np.arange(0,10,1),'Comfort Perception')
comfort['very_uncomfortable'] = fuzz.gaussmf(comfort.universe, 0, 2)
comfort['uncomfortable'] = fuzz.gaussmf(comfort.universe, 5, 2)
comfort['comfortable'] = fuzz.gaussmf(comfort.universe, 9, 1)
comfort.view()
```

## Rules

```python
rule0 = ctrl.Rule(antecedent=( (t_amb['very_cold'] & t_skin['cold_nociceptor']) |
                               (t_amb['very_hot'] & t_skin['heat_nociceptor']) |
                               (t_amb['cold'] & t_skin['cold_nociceptor']) |
                               (t_amb['hot'] & t_skin['heat_nociceptor']) |
                               (force['high']) 
                             ), consequent=comfort['very_uncomfortable'], label='Rule Very Uncomfortable')
rule0.view()
```
